## Add new value at row level

```python
df.head()
```

```python
def pclass_covert(x):
    
    if x == 1:
        return 'First'
    elif x == 2:
        return 'Second'
    elif x == 3:
        return 'Third'
    
#     x_dict = {1 : 'First', 2 : 'Second', 3 : 'Third'}
#     return x_dict[x]

df['Pclass_coverted'] = df["Pclass"].apply(pclass_covert)

df.head()
```

```python
sql_query = """select *, case 
    when Pclass = 1 then 'First'
    when Pclass = 2 then 'Second'
    when Pclass = 3 then 'Third'
    end as Pclass_coverted from titanic"""

cur.execute(sql_query)

for i in (cur.fetchall()):
    print(i)
    
    if i[0] > 6:
        break
```

![](./joins.jpg)

```python
a_df = pd.DataFrame({'A' : [1,2,3,4], 'B' : ['AA', 'AA', 'BB', 'CC']})
b_df = pd.DataFrame({'A' : [3,4,5,6], 'C' : ['AA', 'AA', 'BB', 'CC']})
```

```python
a_df.to_sql("a_data_frame", conn, if_exists='replace', index=False)
b_df.to_sql("b_data_frame", conn, if_exists='replace', index=False)
```

#### Inner Join

```python
pd.merge(a_df, b_df, on='A', how='inner')
```

```python
sql_query = """
select 
    *
from
    a_data_frame as a_tbl
inner join
    b_data_frame as b_tbl
on
    a_tbl.A = b_tbl.A
    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```

#### Left Join

```python
pd.merge(a_df, b_df, on='A', how='left')
```

```python
sql_query = """
select 
    *
from
    a_data_frame as a_tbl
left join
    b_data_frame as b_tbl
on
    a_tbl.A = b_tbl.A
    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```

#### Right Join

```python
pd.merge(a_df, b_df, on='A', how='right')
```

```python
sql_query = """
select 
    *
from
    a_data_frame as a_tbl
right join
    b_data_frame as b_tbl
on
    a_tbl.A = b_tbl.A
    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```

```python
sql_query = """
select 
    *
from
    b_data_frame as b_tbl
left join
    a_data_frame as a_tbl
on
    a_tbl.A = b_tbl.A
    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```

#### Outer Join

```python
pd.merge(a_df, b_df, on='A', how='outer')
```

```python
sql_query = """

    select 
        *
    from
        b_data_frame as b_tbl
    left join
        a_data_frame as a_tbl
    on
        a_tbl.A = b_tbl.A

union

    select 
        *
    from
        a_data_frame as a_tbl
    left join
        b_data_frame as b_tbl
    on
        a_tbl.A = b_tbl.A


    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```

#### Cross Join

```python
pd.merge(a_df, b_df, how='cross')
```

```python
sql_query = """
select 
    *
from
    a_data_frame as a_tbl
cross join
    b_data_frame as b_tbl
    """
cur.execute(sql_query)

for j, i in enumerate(cur.fetchall()):
    print(i)
```
