# Line Plots

```python
# importing required libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#making data
x = np.array([1,4,7])
y = np.array([2,5,2])

plt.plot(x,y,
         marker = 'o', 
         linestyle='-.',
         linewidth=10,
         markersize=40,
         color = 'red')   # line plot
plt.show()
```

```python
#making data
x = np.array([1,4,7])
y = np.array([2,5,2])

plt.plot(x,y,
         marker = 'o', 
         ls='-.',
         lw=10,
         ms=40,
         color = 'red')   # line plot
plt.show()
```

```python
#making data
x = np.array([1,4,7])
y = np.array([2,5,2])

plt.plot(x,y,"o:g", lw = 3, ms = 10,mec='red',mfc='#a2b8a7')   # mec = markeredgecolor
plt.show()
```

```python
#making data
x = np.array([1,4,7])
y1 = np.array([2,5,2])
y2 = [5,2,5]

plt.figure(figsize=(10,7))
plt.plot(x,y1,"o:g", lw = 3, ms = 10,mec='red',mfc='#a2b8a7',label='green')
plt.plot(x,y2,"+-r", lw = 3, ms = 10,mec='green',mfc='hotpink', label='red')
plt.legend(loc = 0)
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()
```

# Bar Plots

```python
fruits = ["Apple","Banana",'Cherry',"Berry"]
counts = [10,12,30,20]

plt.bar(x=fruits,height=counts, color='red')
plt.xlabel("fruits")
plt.ylabel("counts of fruits")
plt.title("Bar Chart", size = 20)
plt.show()
```

```python
fruits = ["Apple","Banana",'Cherry',"Berry"]
counts = [10,12,30,20]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(x = i, y = y[i], s=y[i], ha = 'center')

plt.bar(x=fruits,height=counts, color=['red','yellow','blue','hotpink'])
plt.xlabel("fruits")
plt.ylabel("counts of fruits")
plt.title("Bar Chart", size = 20)
plt.xticks(rotation=45)
addlabels(fruits, counts)

plt.show()
```

```python
fruits = ["Apple","Banana",'Cherry',"Berry"]
counts = [10,12,30,20]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(y = i, x = y[i], s=y[i], ha = 'center')

plt.barh(y=fruits,width=counts, color=['red','yellow','blue','hotpink'])
plt.ylabel("fruits")
plt.xlabel("counts of fruits")
plt.title("Bar Chart", size = 20)
plt.yticks(rotation=45)
addlabels(fruits, counts)

plt.show()
```

```python
fruits = ["Apple","Banana",'Cherry',"Berry"]
counts = [10,12,30,20]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(x = i, y = y[i], s=y[i], ha = 'center')

plt.bar(x=fruits,height=counts, color=['red','yellow','blue','hotpink'], width=0.5)
plt.xlabel("fruits")
plt.ylabel("counts of fruits")
plt.title("Bar Chart", size = 20)
plt.xticks(rotation=45)
addlabels(fruits, counts)

plt.show()
```

# Pie Charts
- can be very useful for ratios/proportions

```python
x = [1,2,3,4]

plt.pie(x, colors = ['black','hotpink','yellow','red'])
plt.title('Pie Chart')
plt.show()
```

```python
counts = [10,25,30,50]
name = ['Apple','Banana','Cherry','Berry']

plt.figure(figsize=(5,5),dpi=100)
plt.pie(x=counts,explode=[0.2,0.3,0.1,0.2],labels=name,autopct="%.2f",shadow = True)
plt.legend(loc=2)
plt.title('Pie Chart')
plt.show()
```

# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,8,1,10]

plt.figure(figsize = (15,7))
plt.subplot(1,3,1)
plt.plot(x,y)

plt.subplot(1,3,2)
plt.bar(x,y)

plt.subplot(1,3,3)
plt.pie(x,labels=y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,8,1,10]

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.bar(x,y)

plt.show()
```

```python
x = [1,2,3,4]
y = [2,8,1,10]

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x,y)
ax2.bar(x,y)
```

# Stacked Plots

```python
round1 = [10,15,12,20]
round2 = [5,8,17,12]

np.array(round1)+np.array(round2)
```

```python
teams = ['a','b','c','d']
round1 = [10,15,12,20]
round2 = [5,8,17,12]
round3 = [2,4,7,9]

plt.bar(teams,round1)
plt.bar(teams,round2, alpha=0.5, color = "green", bottom=round1)
plt.bar(teams,round3,bottom=np.array(round1)+np.array(round2))
```

```python
teams = ['a','b','c','d']
round1 = [10,15,12,20]
round2 = [5,8,17,12]
round3 = [2,4,7,9]

plt.bar(teams,round1,label="round1")
plt.bar(teams,round2,alpha = 0.8, bottom = round1, label ="round2")
plt.bar(teams,round3,alpha=0.8,bottom=np.array(round1)+np.array(round2),color="red",label="round3")
plt.legend()
plt.xlabel("Teams playing ---->")
plt.ylabel("Scores ---->")
plt.title("Match Scores", size = 20, color = "red")
plt.show()
```
