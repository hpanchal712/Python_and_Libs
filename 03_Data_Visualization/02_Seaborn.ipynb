# Seaborn
- Seaborn is also a visualization library built on top of matplotlib.

```python
# import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')

# supress the warnings
import warnings
warnings.filterwarnings('ignore')
```

# Bar Plot

```python
xvalues = ['a','b','c','d']
yvalues = [10,11,17,15]
gender = ['m','f','f','m']

sns.barplot(x=xvalues,y=yvalues)
plt.xlabel('Employee Names')
plt.ylabel('scores')
plt.title("Seaborn Bar Chart Example")
plt.show()     # you can also use semi-colon in place of plt.show()
```

```python
xvalues = ['a','b','c','d']
yvalues = [10,11,17,15]
gender = ['m','f','f','m']

sns.barplot(x=xvalues,y=yvalues,hue=gender)
plt.xlabel('Employee Names')
plt.ylabel('scores')
plt.title("Seaborn Bar Chart Example")
plt.show()     # you can also use semi-colon in place of plt.show()
```

```python
# using iris data
iris = sns.load_dataset('iris')

sns.barplot(x = 'species',
            y = 'sepal_length',
            data = iris)
plt.show()
```

```python
# alternate code

sns.barplot(x = iris['species'],
            y = iris['sepal_length'],
            )
plt.show()
```

# Histograms

```python
# Vertical histogram
plt.figure(figsize=(7,5),dpi=100)
titanic = sns.load_dataset('titanic')
sns.histplot(x="age", data = titanic)
plt.show()
```

```python
# Horizontal Histogram
plt.figure(figsize=(7,5),dpi=100)
titanic = sns.load_dataset('titanic')
sns.histplot(y="age", data = titanic, bins=10)
plt.show()
```

```python
# with hist and kde parameters

sns.distplot(x = titanic['age'],hist=True, kde=True,bins=10,color='red');
```

```python
plt.bar(x=df['days'],height=df['value1'])
plt.bar(x=df['days'],height=df['value2'],alpha=0.6,bottom=df['value1'])
plt.plot(df['total'])
```

```python
x = np.arange(len(df['days']))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df['value1'], width, label='Value1')
rects2 = ax.bar(x + width/2, df['value2'], width, label='Value2')
rects3 = ax.plot(x, df['total'], label='total')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.legend()

fig.tight_layout()

plt.show()
```

# Joint Plot

```python
sns.jointplot(x=titanic.age,
              y=titanic.fare);
```

```python
sns.jointplot(x=titanic.age,
              y=titanic.fare,
              kind='hex');
```

**Different values of kind parameterL**

kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }

```python
sns.jointplot(x=titanic.age,
              y=titanic.fare,
              kind='reg',
              color='red');
```

```python
# with hue parameter

sns.jointplot(x=titanic.age,
              y=titanic.fare,
              kind='scatter',
              hue=titanic.sex);
```

```python
# using iris data 
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_length',
              y='sepal_width',
              kind='scatter',
              hue='species',
              data = iris);
```

```python
# using iris data 
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_length',
              y='sepal_width',
              kind='scatter',
              hue='species',
              data = iris);
```

# Pairplot - Multivariate Analysis
- it takes all the numerical attributes from the dataframe.

```python
# using iris data

sns.pairplot(iris, 
             hue = 'species', 
             kind = 'scatter',
             diag_kind = 'hist')
plt.show()
```

```python
# display the corner figure 

sns.pairplot(iris, 
             hue = 'species', 
             kind = 'kde',
             corner=True,
             diag_kind = 'hist')
plt.show()
```
